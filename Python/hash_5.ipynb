```python
import time
import numpy as np
```

```python
plays=list(np.random.randint(1,1000,1000000))
genres=list(map(str,plays))
```

```python
def solution(genres, plays):
    
    dic_genres={}
    dic_plays={}
    answer=[]

    for i in range(len(genres)):
        if genres[i] in dic_genres:
            dic_genres[genres[i]].append((plays[i],i))
            dic_plays[genres[i]]+=plays[i]

        else:
            dic_genres[genres[i]]=[(plays[i],i)]
            dic_plays[genres[i]]=plays[i]

    dic_plays=sorted(dic_plays.items(), key=(lambda x:x[1]), reverse=True)

    start=time.time()
    for x,y in dic_plays:
        if len(dic_genres[x])==1:
            answer.append(dic_genres[x][1])

        else:
            dic_genres[x].sort(key=lambda x: x[0], reverse=True)
            answer.extend([dic_genres[x][0][1],dic_genres[x][1][1]])
    runtime=time.time()-start
    print(runtime)      
           
    return answer
```

```python
solution(genres, plays)
```

```python
def solution(genres, plays):
    
    dic_genres={}
    dic_plays={}
    answer=[]

    for i in range(len(genres)):
        if genres[i] in dic_genres:
            dic_genres[genres[i]].append((plays[i],i))
            dic_plays[genres[i]]+=plays[i]

        else:
            dic_genres[genres[i]]=[(plays[i],i)]
            dic_plays[genres[i]]=plays[i]

    dic_plays=sorted(dic_plays.items(), key=(lambda x:x[1]), reverse=True)

    start=time.time()
    for x,y in dic_plays:
        genre=dic_genres[x]
        genre=sorted(genre, key=lambda x: x[0], reverse=True)

        cnt=0
        for g in genre:
            answer.append(g[1])
            cnt+=1
            if cnt==2:
                break
    runtime=time.time()-start
    print(runtime)    
            
    return answer
```

```python
solution(genres, plays)
```
